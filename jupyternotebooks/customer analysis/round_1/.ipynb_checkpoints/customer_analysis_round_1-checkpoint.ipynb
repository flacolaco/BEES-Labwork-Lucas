```python
# Import panda and numpy libraries

import numpy as np
import pandas as pd
```

```python
# Import csv file1

file1 = pd.read_csv('file1.csv')
```

```python
# Snapshot of data frame1

file1.head()
```

```python
# Show data frame shape:

file1.shape
```

```python
# Modify some headers

file1=file1.rename(columns={'Customer':'id','ST':'state',})
```

```python
# Created empty list to to change all column headers to lower case

column_names1 = []
for i in range(len(file1.columns)):
    column_names1.append(file1.columns[i].lower())
```

```python
# Change column names of Data Frame 1 to new column names

file1.columns = column_names1
```

```python
# Import csv file2

file2 = pd.read_csv('file2.csv')
```

```python
# Snapshot of data frame2

file2.head()
```

```python
# Show data frame shape:

file2.shape
```

```python
# Modify headers to match headers from Data Frame 1

file2=file2.rename(columns={'Customer':'id','ST':'state',})
```

```python
# Created empty list to to change all column headers to lower case

column_names2 = []
for i in range(len(file1.columns)):
    column_names2.append(file2.columns[i].lower())
column_names2
```

```python
# Change column names of Data Frame 2 to new column names

file2.columns = column_names2
```

```python
# Import csv file3

file3 = pd.read_csv('file3.csv')
```

```python
# Snapshot of data frame3

file3.head()
```

```python
# Show data frame shape:

file3.shape
```

```python
# Modify headers to match headers from Data Frame's 1 & 2

file3=file3.rename(columns={'Customer':'id'})
```

```python
# Created empty list to to change all column headers to lower case

column_names3 = []
for i in range(len(file1.columns)):
    column_names3.append(file3.columns[i].lower())
```

```python
# Change column names of Data Frame 3 to new column names

file3.columns = column_names3
```

```python
# Create Data Frame where column headers are matching column_names 1 (which are also matching 2 and 3)

data = pd.DataFrame(columns = column_names1)
```

```python
# Merge all three Data Frame´s into the newly creted one

data=pd.concat([data,file1,file2,file3],axis=0)
```

```python
# Check newly created Data Frame´s column structure

data.head()
```

```python
# Check newly created Data Frame´s shape (total rows and columns matches the sum of all three other dataframes)

data.shape
```

```python
# Checking for Data Types, Index and Null values per attribute

data.info()

# Categorical columns: state, gender, education, policy type and vehicle class.
```

```python
# Index reset

data.reset_index(drop=True, inplace=True)
```

```python
# Show numerical data: right now only income, monthly premium auto and total claim

data._get_numeric_data()

# Note: customer lifetime value should be numerical!
```

### Column meanings:

    1. id = unique customer identifier
    2. state = state in the us where the customer lives
    3. gender = gender of the customers
    4. education = highest level of education achieved by the customer
    5. customer lifetime value = how much the customer has spent with the company since their first purchase.
    6. income = total yearly income of the customer?
    7. monthly premium auto = monthly amount the customer pays for premium services?
    8. number of open complaints = total currently unresolved complaints in date format?
    9. policy type = customer´s type of car insurance policy
    10. vehicle class = type of vehicle for which the customer has his policy?
    11. total claim amount = total amount claimed by other insurance companies?

```python
# Delete column education and number of open complaints:

data = data.drop(['education','number of open complaints'], axis = 1)
```

```python
# Convert column customer lifetime value to numeric:

data['customer lifetime value']=pd.to_numeric(data['customer lifetime value'],errors='coerce')
```
