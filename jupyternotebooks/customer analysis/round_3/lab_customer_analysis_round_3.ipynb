```python
# Import different libraries for EDA:

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Open file and store it in DataFrame "mca3":

mca3 = pd.read_csv('marketing_customer_analysis.csv')
```

```python
# Snapshot of table columns and first 5 records:

mca3.head()
```

```python
# Show DataFrame info

mca3.info()
```

```python
# DataFrame description:

mca3.describe()
```

```python
mca3['Response'].value_counts().plot(kind = 'bar', color='orange');
```

```python
# Plot of the rate of the Response types by Sales Channel.

plt.figure(figsize=(8,4))
sns.countplot(x='Sales Channel', hue='Response', data=mca3)
plt.ylabel('Count of Response type')
plt.title('Response type by Sales Channel')
plt.show()
```

```python
# Boxplot of Total Claim Amount by Response type. 

sns.boxplot(x="Total Claim Amount", y="Response",data=mca3);
```

```python
# Distribution Plot of Total Claim Amount by Response type. 

sns.catplot(x="Total Claim Amount", y="Response", hue='Response', data=mca3);
```

```python
# Boxplot of Income by Response type. 

sns.boxplot(x="Income", y="Response",data=mca3);
```

```python
# Distribution Plot of Income by Response type. 

sns.histplot(x='Income', hue='Response', data=mca3);
```

```python
# Scatterplot between Total Claim Amount and Income:

sns.jointplot(x=mca3['Income'], y=mca3['Total Claim Amount']);
```

```python
# Scatterplot between Total Claim Amount and Income(2):

sns.regplot(y='Income', x='Total Claim Amount', data=mca3)
```
